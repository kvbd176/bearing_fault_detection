## Import Libraries

```python
import numpy as np
import pandas as pd

from scipy.stats import kurtosis, skew

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score

import joblib
```

## Load Datasets

```python
or12=pd.read_csv("OR007@12_3.csv",header=None)
or6=pd.read_csv("OR007@6_3.csv",header=None)
or3=pd.read_csv("OR007@3_3.csv",header=None)
ir=pd.read_csv("IR007_3.csv",header=None)
ball=pd.read_csv("B007_3.csv",header=None)
```

```python
datasets={
    "OR@12":or12,
    "OR@6":or6,
    "OR@3":or3,
    "inner_race":ir,
    "ball_fault":ball
}
```

```python
for df in datasets.values():
    df.columns=["signal"]
```

## Feature Extraction

```python
window_size=1000
features=[]
fault_labels=[]
```

```python
for fault, df in datasets.items():
    signal=df["signal"].values
    for i in range(0, len(signal)-window_size, window_size,1):
        window=signal[i:i+window_size]
        mean=np.mean(window)
        std=np.std(window)
        rms=np.sqrt(np.mean(window**2))
        maxi=np.max(window)
        mini=np.min(window)
        kurt=kurtosis(window)
        sk=skew(window)
        features.append([mean,std,rms,maxi,mini,kurt,sk])
        fault_labels.append(fault)
feature_names=["mean","std","rms","max","min","kurtosis","skewness"]
```

```python
X=pd.DataFrame(features,columns=feature_names)
fault_labels=np.array(fault_labels)
```

## Calculate Medians

```python
fault_medians={}
for fault in np.unique(fault_labels):
    fault_medians[fault]=X[fault_labels==fault]["rms"].median()
print(fault_medians)
```

## Create Severity Labels

```python
severity_labels=[]
for i in range(len(X)):
    fault=fault_labels[i]
    current_rms=X.iloc[i]["rms"]
    median_rms=fault_medians[fault]
    if current_rms<median_rms*0.9:
        severity_labels.append("Low")
    elif current_rms<=median_rms*1.1:
        severity_labels.append("Medium")
    else:
        severity_labels.append("High")
severity_labels=np.array(severity_labels)
```

## Scale features

```python
import joblib
scaler=joblib.load("scaler.pkl")
```

```python
X_scaled=scaler.transform(X)
X_scaled_df=pd.DataFrame(
    X_scaled,
    columns=feature_names
)
```

## Train-test split

```python
X_train,X_test,y_train,y_test=train_test_split(
    X_scaled_df,
    severity_labels,
    test_size=0.2,
    random_state=42,
    stratify=severity_labels
)
```

## Train Model

```python
severity_model=KNeighborsClassifier(n_neighbors=5)
severity_model.fit(X_train,y_train)
print("Severity model trained successfully!")
```

## Model Evaluation

```python
y_pred=severity_model.predict(X_test)
print("Accuracy:",accuracy_score(y_test, y_pred)*100)
print(classification_report(y_test,y_pred))
```

## Save the files

```python
joblib.dump(severity_model,"severity_model.pkl")
print("Files saved successfully!")
```
